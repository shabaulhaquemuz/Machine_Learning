```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("credit_scoring.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df,columns=["Gender","Marital Status","Education Level"])
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
df["Employment Status"]=oe.fit_transform(df[["Employment Status"]])
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Type of Loan"]=oe.fit_transform(df[["Type of Loan"]])
df.head(2)
```

```python
df.astype(int)
```

```python
x=df.drop(columns="Type of Loan")
y=df["Type of Loan"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rd=RandomForestClassifier(n_estimators=200,random_state=42)
rd.fit(x_train,y_train)
```

```python
y_pred=rd.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)
```

```python
param_grid = [
    {'penalty':['l1','l2','elasticnet','none'],
    'C' : np.logspace(-4,4,20),
    'solver': ['lbfgs','newton-cg','liblinear','sag','saga'],
    'max_iter'  : [10,50,100,150,500,1000,2000,3000]
}
]
```

```python
from sklearn.model_selection import GridSearchCV
gridsearch = GridSearchCV(lr,param_grid = param_grid, cv = 9, verbose=True,n_jobs=-1)
gridsearch
```

```python
bestgridsearch = gridsearch.fit(x_train,y_train)
bestgridsearch.best_estimator_
```

```python
print(f'Accuracy : {bestgridsearch.score(x,y):.3f}')
```
