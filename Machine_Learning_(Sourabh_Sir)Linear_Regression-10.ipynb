# Insurance Dataset

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("insurance.csv")
df.head(3)
```

```python
df =pd.get_dummies(df,columns=['sex','smoker','region'],drop_first=True)
df.head(3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=['charges'])
y = df['charges']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()  #Creates a Linear Regression model object.
lr.fit(x_train,y_train)  #Fits (trains) the model on the training data to learn coefficients.
```

```python
y_pred = lr.predict(x_test)

# Uses the trained model to predict charges for test data (x_test).
# y_pred holds the predicted values.

# x_test â†’ model input (features like age, BMI, smoker, etc.)
# y_pred â†’ modelâ€™s predicted charges for those inputs
# y_test â†’ actual charges from the dataset (ground truth)

# x_test only has features (age, BMI, smoker, region, etc.) but not charges.
# The model uses those features to predict y_pred (charges).
# Then we compare these predicted charges with the actual charges (y_test).
# Finally, we calculate RÂ² (or other metrics) to see how well the model performed.
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)

# RÂ² score basically tells us how well the model is performing at predicting the target.
# ðŸ‘‰ Higher RÂ² = better performance, lower RÂ² = poor performance.
#RÂ² score tells us how well the modelâ€™s predictions(y_pred) match the real values(y_test) â€” higher means better.
```

```python
# | **RÂ² Score Range** | **Interpretation**                                                        |
# | ------------------ | ------------------------------------------------------------------------- |
# | **1.0 (100%)**     | Perfect prediction â†’ model predicts actual values exactly                 |
# | **â‰¥ 0.8 (80%+)**   | Very good prediction â†’ model explains most of the variance                |
# | **0.5 â€“ 0.8**      | Moderate prediction â†’ model explains some variance, usable but not great  |
# | **0.0 â€“ 0.5**      | Poor prediction â†’ model barely explains the data, weak fit                |
# | **< 0.0**          | Very poor prediction â†’ worse than just predicting the average of `y_test` |

#Formula:
# R^2=1-(Unexpalined Variance/Total Variance)
# UnexplainedÂ VarianceÂ (RSS)=âˆ‘(yiâ€‹âˆ’yi^â€‹)^2
# TotalÂ VarianceÂ (TSS)=âˆ‘(yiâ€‹âˆ’yË‰â€‹)^2

```

# BMW Sales Dataset

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("BMW sales data (2010-2024).csv")
df.head(3)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,StandardScaler,MinMaxScaler
from sklearn.impute import SimpleImputer
```

```python
numeric = ['year','Engine_Size_L','Mileage_KM',"Price_USD"]
categoric = ['Model','region','Color','Fuel_Type','Transmission','Sales_Classification']
```

```python
numeric_Transfromer = Pipeline(steps=[
    ('Impute',SimpleImputer(strategy='mean')),
    ('Standarize',StandardScaler()),
    ('Normalize',MinMaxScaler())
])

categoric_Transformer_1 = Pipeline(steps = [
        ('Impute',SimpleImputer(strategy='most_frequent')),
        ('Encode',LabelEncoder())

])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('Numeric',numeric_Transfromer,numeric),
    ('Categoric',categoric_Transformer_1,categoric)
])
```

```python
x = df.drop(columns=['Sales_Volume'])
y = df['Sales_Volume']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train = pd.get_dummies(x_train, drop_first=True)
x_test = pd.get_dummies(x_test, drop_first=True)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
