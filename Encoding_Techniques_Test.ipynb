```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("userbehaviour.csv")
```

```python
print("before encoding:")
df.head(5)
```

```python
df=df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Create LabelEncoder objects
status_encoder=LabelEncoder()


#Apply label encoding
df["Status"]=status_encoder.fit_transform(df["Status"])

print("after label encoding:")
print(df)
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler
scaler_standard = StandardScaler()
df_standardized = df.copy()
df_standardized[df.columns] = scaler_standard.fit_transform(df)

df_standardized.to_csv("userbehaviour_standardized.csv", index=False)
print("after standardization:")
print(df_standardized)
```

```python
# Normalization
from sklearn.preprocessing import MinMaxScaler
scaler_minmax = MinMaxScaler()
df_normalized = df.copy()
df_normalized[df.columns] = scaler_minmax.fit_transform(df)


df_normalized.to_csv("userbehaviour_normalized.csv")

print("after normalization:")
print(df_normalized)
```

```python
#Ordinal Encoding
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df = pd.read_csv("userbehaviour.csv")

df = df.dropna()

ordinal_encoder = OrdinalEncoder()
df["Status"] = ordinal_encoder.fit_transform(df[["Status"]])

df.to_csv("userbehaviour_ordinal.csv")

print("after ordinal encoding:")
print(df.head())


```
