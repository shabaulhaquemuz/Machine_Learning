```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("tips.csv")
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df,columns=["sex","smoker","day","time"])
```

```python
x=df.drop(columns="total_bill")
y=df["total_bill"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import Lasso

param_dist = {'alpha': [0.1, 1.0, 10.0, 100.0]}
lasso = Lasso()
random_search = RandomizedSearchCV(lasso, param_dist, n_iter=2, cv=3)
random_search.fit(x_train, y_train)
```

```python
y_pred_new = random_search.predict(x_test)
r2_score(y_test, y_pred_new)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

param_grid = {'alpha': [0.1, 1.0, 10.0, 100.0]}
ridge = Ridge()
grid_search = GridSearchCV(ridge, param_grid, cv=2)
grid_search.fit(x_train, y_train)
```

```python
print("accuracy for grid search:")
y_pred_grid = grid_search.predict(x_test)
r2_score(y_test, y_pred_grid)
```
