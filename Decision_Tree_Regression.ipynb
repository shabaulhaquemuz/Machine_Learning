```python
#How decision tree algorithm works?
#The algorithm starts at the root node and recursingly splits the data into smaller subsets based on specific features and their values.
#The goal of each split is to create child notes that are more homogeneous in  terms of the target variable.

#To determine the best split at each node, the algorithm typically uses a matrix like MSE or MAE. It evaluates potenital split points for each feature
#and choses the scale that minimize the error in the resulting child node. This process aims to reduce the variance of the target variable within each
# subset.
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("insurance.csv")
df.head(3)
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df,columns=["sex","smoker","region"])
```

```python
df.head(3)
```

```python
df.astype(int)
```

```python
x=df.drop(columns="charges")
y=df["charges"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
